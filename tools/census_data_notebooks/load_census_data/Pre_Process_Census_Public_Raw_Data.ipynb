## notes

```python
df_notes = pd.read_csv("notes22.csv")
```

```python
#df_notes.head()
```

```python
df_notes.dtypes
```

```python
df_notes['TITLE'] = df_notes['TITLE'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_notes['CONTENT'] = df_notes['CONTENT'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
df_notes.to_csv("notes22_processed.csv")
```

```python
df_cfda = pd.read_csv("cfda22.csv", low_memory=False, \
                      dtype={'CFDA':object, 'STATECLUSTERNAME': object, 'OTHERCLUSTERNAME':object})
```

```python
#df_cfda.head()
```

```python
df_cfda.dtypes
```

```python
df_cfda['AWARDIDENTIFICATION'] = df_cfda['AWARDIDENTIFICATION'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_cfda['FEDERALPROGRAMNAME'] = df_cfda['FEDERALPROGRAMNAME'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_cfda['CFDAPROGRAMNAME'] = df_cfda['CFDAPROGRAMNAME'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
df_cfda.to_csv("cfda22_processed.csv")
```
