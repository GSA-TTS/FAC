### Create subset for team by excluding tribal data

Note: Subset df_elecauditheader using (SUPPRESSION_CODE = Null) & (ENTITY_TYPE in ['Non-profit','Local government', 'State', 'Institution of higher education (IHE)'] & AUDITYEAR in [2016,2017,2018,2019,2020,2021,2022]

```python
df_elecauditheader_sub = df_elecauditheader[df_elecauditheader['AUDITYEAR'].isin(['2016','2017','2018','2019','2020','2021','2022']) &
                                            (df_elecauditheader['SUPPRESSION_CODE'].isna()) &
                                            df_elecauditheader['ENTITY_TYPE'].isin([
                                                'Non-profit','Local government','Institution of higher education (IHE)',
                                                'State','NON-PROFIT','Non-Profit','Local Government','LOCAL GOVERNMENT',
                                                'INSTITUTION OF HIGHER EDUCATION (IHE)','STATE'])
                                            ]
df_elecauditheader_sub.shape
```

```python
df_elecauditheader_sub["ENTITY_TYPE"].value_counts(dropna=False)
```

```python
df_elecauditheader_sub["AUDITYEAR"].value_counts(dropna=False)
```

```python
df_elecauditheader_sub["SUPPRESSION_CODE"].value_counts(dropna=False)
```
