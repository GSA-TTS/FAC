## Connect to Postgres database

```python
url = "postgresql://postgres:1234@localhost:5432/postgres"

connection = psycopg2.connect(url)
connection.autocommit = True
cursor = connection.cursor()

cursor.execute("SELECT version();")
version = cursor.fetchone()
print(version)
```

## Create table in Postgres database

```python
# drop table if it already exists
cursor.execute('drop table if exists suk_cfda_agencies')
  
sql = '''CREATE TABLE suk_cfda_agencies(AGENCYNUM int,AGENCYNAME char(512),I_9 char(8), III_9 char(8), III_10A char(8));'''
cursor.execute(sql)

connection.close()
```

## Update table in Postgres database 

```python
url = "postgresql://postgres:1234@localhost:5432/postgres"
engine = create_engine(url)
df.to_sql('suk_cfda_agencies', con=engine, if_exists= 'replace')
```

```python
url = "postgresql://postgres:1234@localhost:5432/postgres"

connection = psycopg2.connect(url)
connection.autocommit = True
cursor = connection.cursor()
# fetching all rows
sql1='''select * from suk_cfda_agencies;'''
cursor.execute(sql1)
for i in cursor.fetchall():
    print(i)
connection.close()
```
