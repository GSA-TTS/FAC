### This notebook is used to subset ELECAUDITS, ELECEINS, ELECPAS, ELECAUDITFINDINGS, ELECPASSTHROUGH, ELECNOTES, ELECFINDINGSTEXT, ELECCAPTEXT, ELECUEIS using the ELECAUDITHEADER.csv that was created with subset_ELECAUDITHEADER... notebook.

### Set year to 2022 and run.
### Repeat for other years 2016, 2017, 2018, 2019, 2020, 2021, one year at a time.

```python
import pandas as pd
```

# Set year

```python
global year
year ='2022'
```

## ELECAUDITHEADER

```python
df_elecauditheader = pd.read_csv(year + "/ELECAUDITHEADER.csv")
df_elecauditheader.head()
```

```python
df_elecauditheader.shape
```

## ELECAUDITS

```python
df_elecaudits = pd.read_csv("/Users/sudhaukumar/Documents/Development/Census_raw_data_to_S3/raw_data_for_GFAC/DISSEMINATION-ELECAUDITS-EXPORT.csv", low_memory=False)
df_elecaudits.head()
```

```python
df_elecaudits.columns
```

```python
df_elecaudits.shape
```

### Subset ELECAUDITS

```python
df_elecaudits_sub = df_elecaudits[(df_elecaudits['AUDITYEAR'] == year) &
                                  (df_elecaudits['DBKEY'].isin(list(df_elecauditheader['DBKEY'].astype(str)))) &
                                  (df_elecaudits['EIN'].isin(list(df_elecauditheader['EIN'].astype(str))))
                                 ]
df_elecaudits_sub.shape
```

```python
df_elecaudits_sub['AUDITYEAR'].value_counts(dropna=False)
```

```python
df_elecaudits_sub.to_csv(year + "/ELECAUDITS.csv", index=False)
```
